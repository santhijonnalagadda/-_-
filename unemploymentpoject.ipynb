```python
# importing the library

import pandas as pd
```

```python
# loading the dataset

path=r'c:\Users\santh\OneDrive\Documents\Unemployment in India.csv'
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
# remove optional columns

df.columns=df.columns.str.strip()
```

```python
# calculating estimated total population

df['Estimated Employed']=df['Estimated Employed']/(1-df['Estimated Unemployment Rate (%)']/100)
```

```python
# calculating average unemployment rate by region

region_avg= df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)
```
