```python
# ADVERTISING AND ITS IMPACT OF SALES
# importing the libraries

import pandas as pd 
import numpy as np
```

```python
# loading the dataset using pandas

path=r'c:\Users\santh\OneDrive\Documents\Sales Advertising.csv'
df=pd.read_csv(path)
```

```python
df
```

```python
# importing thr libraries for predicting the sales 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
# splitting the dataset into features and target variables

x=df[['TV','Radio','Newspaper']]
y=df[['Sales']]
```

```python
# testing and training the dataset through the algorithm Linear regression

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# evaluating the model using the metrics 

y_pred=model.predict(x_test)
MAE=mean_absolute_error(y_test,y_pred)
MSE=mean_squared_error(y_test,y_pred)
RMSE=np.sqrt(MSE)
print(f"mean absolute error:{MAE}")
print(f"mean squared error:{MSE}")
print(f"root mean squared error:{RMSE}")
```
